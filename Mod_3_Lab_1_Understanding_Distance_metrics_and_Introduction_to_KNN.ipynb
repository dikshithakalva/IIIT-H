### Euclidean Distance

Euclidean distance is the length of the line segment between two points in Euclidean space. It is also known as the Pythagorean distance because it can be calculated using the Pythagorean theorem.


image.png

```python
# Simple 2D example
x_1 = np.array((1, 2))
x_2 = np.array((4, 6))
```

```python
# Naive approach to calculating Euclidean distance (not-vectorized)
sumv = 0
for i in range(len(x_1)):
  sumv += ((x_1[i]-x_2[i])**2)
dist = math.sqrt(sumv)
print(dist)
```

```python
# Libraries such as SciPy provide functions to compute different kinds of distance metrics between points
euclidean_dist = distance.euclidean(x_1, x_2)
print(euclidean_dist)
```

### **Hamming Distance**

Hamming distance is a metric for comparing two binary data strings. While comparing two binary strings of equal length, Hamming distance is the number of bit positions in which the two bits are different.

The Hamming distance between two strings, a and b is denoted as d(a,b).


image.png

```python
str_1 = 'euclidean'
str_2 = 'manhattan'
hamming_dist = distance.hamming(list(str_1), list(str_2)) * len(str_1)
print(hamming_dist)
```

### **Chebyshev Distance**

image.png

The Chebyshev distance calculation, commonly known as the "maximum metric" in mathematics, measures distance between two points as the maximum difference over any of their axis values. In a 2D grid, for instance, if we have two points (x1, y1), and (x2, y2), the Chebyshev distance between is max(y2 - y1, x2 - x1)

### **JACCARD DISTANCE**

image.png

Jaccard distance is commonly used to calculate an nxn matrix for clustering and multidimensional scaling of n sample sets. This distance is a metric on the collection of all finite sets. , since these formulas are not well defined in these cases.

```python
## In the context of binary vectors, let's say:
## a = no. of indices where vector1[i] == 1 == vector2[i]
## b = no. of indices where vector1[i] == 0 and vector2[i] == 1
## c = no. of indices where vector1[i] == 1 and vector2[i] == 0
## d = no. of indices where vector1[i] == 0 == vector2[i]
## Jaccard-distance = 1 - [a/(a+b+c)]

print(distance.jaccard([1, 0, 0], [0, 1, 0]))
print(distance.jaccard([1, 0, 0], [1, 1, 0]))
print(distance.jaccard([1, 0, 0], [1, 1, 1]))
```

```python
def jaccard_similarity(a, b):
    c = a.intersection(b)
    return float(len(c)) / (len(a) + len(b) - len(c))

list1 = ['dog', 'cat', 'rat']
list2 = ['dog', 'cat', 'mouse', 'rabbit']
# The intersection is ['dog', 'cat']
# union is ['dog', 'cat', 'rat', 'mouse, 'rabbit']
words1 = set(list1)
words2 = set(list2)
print("Jaccard similarity is :",jaccard_similarity(words1, words2))
print("Jaccard distance is :", 1-jaccard_similarity(words1, words2))
```

### **KNN**
 K-Nearest Neighbors Algorithm. The k-nearest neighbors algorithm, also known as KNN or k-NN, is a non-parametric, supervised learning classifier, which uses proximity to make classifications or predictions about the grouping of an individual data point.

```python
import random
import scipy.stats as ss
from sklearn.neighbors import KNeighborsClassifier
```

```python
def generate_synth_data(n = 50):
    points = np.concatenate((ss.norm(0, 1).rvs((n, 2)), ss.norm(1, 1).rvs((n, 2))), axis = 0)
    outcomes = np.concatenate((np.repeat(0, n), np.repeat(1, n)))
    return (points, outcomes)

n = 50
pts,tgts = generate_synth_data(n) #generates 100 points
# print(tgts)
plt.figure()
plt.plot(pts[:n, 0], pts[:n, 1], "ro")
plt.plot(pts[n:, 0], pts[n:, 1], "bo")
plt.show()
```

```python
n = 15
test_pts,test_tgts = generate_synth_data(n)
# print(test_tgts)
plt.figure()
plt.plot(test_pts[:n, 0], test_pts[:n, 1], "ro")
plt.plot(test_pts[n:, 0], test_pts[n:, 1], "bo")
plt.show()
```

```python
## TASK - 3

knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(pts, tgts)
our_predictions = knn.predict(test_pts)
print("Prediction Accuracy: ")
print(100 * np.mean(our_predictions == test_tgts))
```

### **SECTION - 2.2: KNN on a Real World dataset**

The Iris flower data set or Fisher's Iris data set is a multivariate data set that consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters

```python
from sklearn.datasets import load_iris
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
# data['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
data['target'] = pd.Series(iris.target)
data.head(10)
```

```python
np.random.seed(123)
indices = np.random.permutation(data.shape[0])
div = int(0.65 * len(indices))
train_idx, test_idx = indices[:div], indices[div:]

train_set, test_set = data.loc[train_idx,:], data.loc[test_idx,:]
test_class = list(test_set.iloc[:,-1])
train_class = list(train_set.iloc[:,-1])
```

```python
def dist_euclidean(X1,X2):
    return distance.euclidean(X1,X2)
```

```python
def dist_cosine(X1,X2):
    return distance.cosine(X1,X2)
```

```python
def dist_manhattan(X1,X2):
    return distance.cityblock(X1,X2)
```

```python
def dist_chebyshev(X1,X2):
    return distance.chebyshev(X1,X2)
```

```python
metrics_list = [dist_euclidean,dist_cosine, dist_manhattan, dist_chebyshev]
for dist_metric in metrics_list:
  knn = KNeighborsClassifier(n_neighbors = 3, metric=dist_metric)
  knn.fit(train_set, train_class)
  our_predictions = knn.predict(test_set)
  print(dist_metric.__name__)
  print("Prediction Accuracy: ")
  print(100 * np.mean(our_predictions == test_class))
  print()
```

### **How are similarity and distance different from each other?**

1. Similarity in ML often refers to how alike two data points are in terms of their features. It's a measure of resemblance or likeness between them. Common similarity metrics include cosine similarity, Jaccard similarity, and Euclidean similarity.

2. Distance, on the other hand, is a measure of how far apart two data points are in the feature space. It's like a numerical value representing the "gap" between them. Common distance metrics include Euclidean distance, Manhattan distance, and Hamming distance.

3. These metrics play a crucial role in various machine learning tasks. Cosine similarity, for example, is handy when dealing with text data, measuring the cosine of the angle between two vectors. Jaccard similarity is often used for sets, like comparing the similarity between two documents based on the presence of certain words.

4. On the distance side, Euclidean distance is like a straight-line distance between points, while Manhattan distance considers only horizontal and vertical movements. And Hamming distance is useful for comparing binary strings.

### **Are there any conditions for a particular distance to be considered a "distance metric"?**

These properties ensure that the distance measure behaves intuitively and consistently. Here are the key conditions for a distance metric:

1. **Non-negativity**: The distance between any two points should always be non-negative.

2. **Identity of indiscernibles**: The distance between two points is zero if and only if the points are identical.

3. **Symmetry**: The distance from point A to point B should be the same as the distance from B to A.

4. **Triangle inequality**: The distance from A to C should be less than or equal to the sum of the distances from A to B and B to C. In other words, taking a detour shouldn't make the distance shorter.

These conditions ensure that the distance measure is reliable and reflects the intuitive notion of distance. Many common distance metrics, such as Euclidean distance, Manhattan distance, and Minkowski distance, adhere to these properties.
